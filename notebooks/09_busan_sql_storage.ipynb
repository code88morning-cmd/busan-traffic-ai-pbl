```python
import sqlite3

conn = sqlite3.connect("data/processed/busan_example.db")
cursor = conn.cursor()
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS weather (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    date TEXT,
    temp REAL,
    humidity REAL
)
""")
conn.commit()
```

```python
cursor.execute("""
INSERT INTO weather (date, temp, humidity)
VALUES (?, ?, ?)
""", ("2026-01-18", -2.2, 63))

conn.commit()
```

```python
data = [
    ("2026-01-19", 0.5, 58),
    ("2026-01-20", 1.2, 55)
]

cursor.executemany("""
INSERT INTO weather (date, temp, humidity)
VALUES (?, ?, ?)
""", data)

conn.commit()
```

```python
cursor.execute("SELECT * FROM weather")
rows = cursor.fetchall()

rows
```

```python
import pandas as pd

df = pd.DataFrame(rows, columns=["id", "date", "temp", "humidity"])
df
```
