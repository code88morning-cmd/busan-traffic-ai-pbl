```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import fetch_california_housing

data = fetch_california_housing()
X = data.data
y = data.target
feature_names = data.feature_names

df = pd.DataFrame(X, columns=feature_names)
df["target"] = y
df.head()
```

```python
df.isna().sum()
```

```python
plt.figure()
plt.hist(df["target"], bins=30)
plt.xlabel("target")
plt.ylabel("count")
plt.title("Target Distribution")
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    df[feature_names].values, df["target"].values, test_size=0.2, random_state=42
)

X_train.shape, X_test.shape
```

```python
X_train, X_val, y_train, y_val = train_test_split(
    X_train, y_train, test_size=0.25, random_state=42
)

X_train.shape, X_val.shape, X_test.shape
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_s = scaler.fit_transform(X_train)
X_val_s   = scaler.transform(X_val)
X_test_s  = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train_s, y_train)

pred_val = model.predict(X_val_s)
pred_val[:5]
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

mae = mean_absolute_error(y_val, pred_val)
rmse = mean_squared_error(y_val, pred_val)

mae, rmse
```

```python
plt.figure()
plt.scatter(y_val, pred_val, s=8)
plt.xlabel("True")
plt.ylabel("Pred")
plt.title("Validation True vs Pred")
plt.show()
```
